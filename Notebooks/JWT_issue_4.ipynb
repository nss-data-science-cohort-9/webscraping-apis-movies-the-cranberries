```python
plt.figure(figsize=(10,6))

sns.regplot(
    data=movies,
    x='Vote_Average',
    y='Revenue',
    scatter_kws={'alpha': 0.4},
)

plt.yscale('log')

# Format y-axis as dollars (even on log scale)
fmt = FuncFormatter(lambda x, _: f'${x:,.0f}')
plt.gca().yaxis.set_major_formatter(fmt)

plt.title('Ratings vs Revenue (Log Scale)')
plt.xlabel('Average Rating')
plt.ylabel('Revenue (USD, log scale)')
plt.show()
```
